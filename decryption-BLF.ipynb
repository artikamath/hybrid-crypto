```
import random
import string
from Crypto.Cipher import Blowfish
import time
begin=time.time()
fk = open('/content/gdrive/My Drive/minor project/sample_key_BLF.txt',"rb")
key_BLF=fk.read()
fin = open('/content/gdrive/My Drive/minor project/sample_encrypted_BLF.txt',"rb")
ciphertext_BLF=fin.read()
cipher_BLF = Blowfish.new(key_BLF)
decryptedtext_BLF=cipher_BLF.decrypt(ciphertext_BLF)
decryptedtext_BLF=decryptedtext_BLF.rstrip()
decryptedtext_BLF=str(decryptedtext_BLF,'utf-8')  #decrypted text
fout=open('/content/gdrive/My Drive/minor project/sample_decrypted_BLF.txt', "w")
fout.write(decryptedtext_BLF)

end=time.time()
print(f"Runtime of the program is {end - begin}")

```
