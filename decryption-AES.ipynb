```
import random
import string
from Crypto.Cipher import AES
import time
begin=time.time()
fk = open('/content/gdrive/My Drive/minor project/sample_key_AES.txt',"rb")
key_AES=fk.read()
fin = open('/content/gdrive/My Drive/minor project/sample_encrypted_AES.txt',"rb")
ciphertext_AES=fin.read()
cipher_AES = AES.new(key_AES)
decryptedtext_AES=cipher_AES.decrypt(ciphertext_AES)
decryptedtext_AES=decryptedtext_AES.rstrip()
decryptedtext_AES=str(decryptedtext_AES,'utf-8')  #decrypted text
fout=open('/content/gdrive/My Drive/minor project/sampledecrypted_AES.txt', "w")
fout.write(decryptedtext_AES)
end=time.time()
print(f"Runtime of the program is {end - begin}")
```
