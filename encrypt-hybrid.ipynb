```
import random
import string
from Crypto.Cipher import Blowfish
from Crypto.Cipher import AES
from Crypto.Random import get_random_bytes

f = open('/content/gdrive/My Drive/minor project/sample-test.txt')

text = f.read() #plain text
plaintext=text.split(" ")
p=int(len(plaintext)*.3)
encrypt_text1=text[0:p]  #text to be encrypted
encrypt_text2=text[p:]

while len(bytes(encrypt_text1, encoding='latin1')) % 16 != 0:  #padding to make text size a multiple of 16
     encrypt_text1 = encrypt_text1 + " "
while len(bytes(encrypt_text2, encoding='latin1')) % 16 != 0:  #padding to make text size a multiple of 16
     encrypt_text2 = encrypt_text2 + " "

key = get_random_bytes(16) 
fk=open('/content/gdrive/My Drive/minor project/sample_key.txt', "ab")
fk.write(key)
cipher_BLF = Blowfish.new(key)
cipher_AES = AES.new(key) 
ciphertext_BLF = cipher_BLF.encrypt(encrypt_text1)
ciphertext_AES = cipher_AES.encrypt(encrypt_text2)

f1 = open('/content/gdrive/My Drive/minor project/sample_encrypted_1.txt', "ab")
f1.write(ciphertext_BLF)
f2 = open('/content/gdrive/My Drive/minor project/sample_encrypted_2.txt', "ab")
f2.write(ciphertext_AES)
f.close()
fk.close()
f1.close()
f2.close()
```
