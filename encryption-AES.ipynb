```
import random
import string
from Crypto.Cipher import AES
from Crypto.Random import get_random_bytes
import time
begin=time.time()
f = open('/content/gdrive/My Drive/minor project/sample.txt')

plaintext = f.read() #plain text
encrypt_text=plaintext  #text to be encrypted
while len(bytes(encrypt_text, encoding='utf-8')) % 16 != 0:  #padding to make text size a multiple of 16
     encrypt_text = encrypt_text + " "
key_AES = get_random_bytes(16) 
fk=open('/content/gdrive/My Drive/minor project/sample_key_AES.txt', "ab")
fk.write(key_AES)
cipher_AES = AES.new(key_AES) 
ciphertext_AES = cipher_AES.encrypt(encrypt_text)
fin = open('/content/gdrive/My Drive/minor project/sample_encrypted_AES.txt', "ab")
fin.write(ciphertext_AES)
f.close()
fk.close()
fin.close()
end=time.time()
print(f"Runtime of the program is {end - begin}")

```
