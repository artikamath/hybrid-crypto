```
import random
import string
from Crypto.Cipher import Blowfish
from Crypto.Cipher import AES

fk = open('/content/gdrive/My Drive/minor project/sample_key.txt',"rb")
key=fk.read()
f1 = open('/content/gdrive/My Drive/minor project/sample_encrypted1.txt',"rb")
f2 = open('/content/gdrive/My Drive/minor project/sample_encrypted2.txt',"rb")
cipher_text1=f1.read()
cipher_text2=f2.read()
cipher_BLF = Blowfish.new(key)
decryptedtext_BLF=cipher_BLF.decrypt(cipher_text1)
decryptedtext_BLF=decryptedtext_BLF.rstrip()
decryptedtext_BLF=str(decryptedtext_BLF,'latin1')  #decrypted text
#fout=open('/content/gdrive/My Drive/minor project/sampledecrypted.txt', "w")
#fout=open('/content/gdrive/My Drive/minor project/sampledecrypted1.txt', "w")
#fout.write(decryptedtext_BLF)

cipher_AES = AES.new(key)
decryptedtext_AES=cipher_AES.decrypt(cipher_text2)
decryptedtext_AES=decryptedtext_AES.rstrip()
decryptedtext_AES=str(decryptedtext_AES,'latin1')  #decrypted text
#fout.write(decryptedtext_AES)

print(decryptedtext_BLF,decryptedtext_AES)
```
