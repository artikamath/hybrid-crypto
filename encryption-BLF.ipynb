```
import random
import string
from Crypto.Cipher import Blowfish
from Crypto.Cipher import AES
from Crypto.Random import get_random_bytes
import time
begin=time.time()
f = open('/content/gdrive/My Drive/minor project/sample.txt')

plaintext = f.read() #plain text
encrypt_text=plaintext  #text to be encrypted
while len(bytes(encrypt_text, encoding='utf-8')) % 16 != 0:  #padding to make text size a multiple of 16
     encrypt_text = encrypt_text + " "
key_BLF = get_random_bytes(10) 
fk=open('/content/gdrive/My Drive/minor project/sample_key_BLF.txt', "ab")
fk.write(key_BLF)
cipher_BLF = Blowfish.new(key_BLF) 
ciphertext_BLF = cipher_BLF.encrypt(encrypt_text)
fin = open('/content/gdrive/My Drive/minor project/sample_encrypted_BLF.txt', "ab")
fin.write(ciphertext_BLF)
f.close()
fk.close()
fin.close()
end=time.time()
print(f"Runtime of the program is {end - begin}")






```
